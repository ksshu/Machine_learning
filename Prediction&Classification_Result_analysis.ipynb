```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/drive/MyDrive/agmr/K S Shubham (1)/6TH SEM/RESULT-ANALYSIS/6th sem result2023(After RV).xlsx")
```

```python
print(df.columns.tolist())
```

```python
df.drop(["Unnamed: 24","Unnamed: 25"],axis=1,inplace=True)
```

```python
df.iloc[8:24]

```

```python
df['RESULT_BINARY'] = df['RESULT'].map({'PASS': 1, 'FAIL': 0})
```

```python
# Features (input for prediction)
X = df[['18EC61', '18EC62', '18EC63', '18EC64', '18EC653', '18ECL66', '18ECL67', '18ECMP68']]



# Targets:
y_total = df['GrandTotal']          # For regression (total marks)
y_result = df['RESULT_BINARY']      # For classification (pass/fail)
```

```python
df=df.iloc[8:24]
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split

# Train/test split (optional)
X_train, X_test, y_total_train, y_total_test = train_test_split(X, y_total, test_size=0.2, random_state=42)
_, _, y_result_train, y_result_test = train_test_split(X, y_result, test_size=0.2, random_state=42)

# Regression model
reg_model = LinearRegression()
reg_model.fit(X_train, y_total_train)

# Classification model
clf_model = LogisticRegression()
clf_model.fit(X_train, y_result_train)
```

```python
# Let's say we have no input and want to predict marks for a new random student
# Use mean values of existing dataset
import numpy as np

x_input = [X[col].mean() for col in X.columns]  # [mean of each subject]
x_input = np.array(x_input).reshape(1, -1)

# Predict subject-wise total (approx)
pred_subjects = x_input.flatten().round()

# Predict grand total
pred_total = reg_model.predict(x_input)[0]

#Predic Percentage
pred_Percentage = (pred_total/600)*100

# Predict pass/fail
pred_result = clf_model.predict(x_input)[0]

print(" Predicted Subject Marks:", pred_subjects)
print(" Predicted Grand Total:", round(pred_total))
print(" Predicted Percentage:", round(pred_Percentage))
print(" Predicted Result:", "PASS" if pred_result == 1 else "FAIL")
```
